```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("CarPrice.csv")
```

```python
df
```

```python
df.drop(['compressionratio','symboling','enginelocation','carbody','boreratio','CarName','peakrpm'],axis=1,inplace=True)
```

```python
df
```

```python
df.drop(['stroke','enginesize','carheight','carlength','carwidth','curbweight','wheelbase','enginetype','fuelsystem'],axis=1,inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.duplicated()
```

```python
df.info()
```

```python
lst=[]
for i in df.columns:
  if df[i].dtype=='object':
    lst.append(i)
```

```python
le=LabelEncoder()
```

```python
for i in lst:
  df[i]= le.fit_transform(df[i])
```

```python
df
```

```python
y=df['price']
```

```python
x=df.drop('price',axis=1)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3)
```

```python
dt=DecisionTreeRegressor()
```

```python
dt.fit(xtrain,ytrain)
```

```python
preds=dt.predict(xtest)
```

```python
root_mean_squared_error(preds, ytest)
```

```python
rf=RandomForestRegressor()
```

```python
rf.fit(xtrain,ytrain)
```

```python
preds=rf.predict(xtest)
```

```python
root_mean_squared_error(preds, ytest)
```
